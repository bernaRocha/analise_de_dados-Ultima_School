# Teorema do limite central


ENUNCIADO

    "Teorema: Para amostras aleatórias simples (X1,...,Xn), retiradas de uma população com média μ e variância σ2 finita, a distribuição amostral da média X¯ aproxima-se, para n grande, de uma distribuição normal, com média μ e variância σ2/n". (Bussab & Morettin, 2003, p. 273-274)

    Em outras palavras, o TLC nos diz que, não importa qual seja a distribuição de uma população, se tirarmos um número N suficientemente grande de amostras dessa população, a distribuição das médias das amostras aproxima-se de uma distribuição normal.

    Além disso, ele nos diz que a médias das médias amostrais aproxima-se da média populacional.



## SIMULAÇÃO

Para visualizarmos este "fenômeno" vamos construir uma simulação com os seguintes passos:

    Simular uma populaçãoo qualquer de 1000 números selecionados aleatoriamente de um intervalo de números inteiros entre 0 e 100.

    Calcular os parâmetros desta população, μ e σ.

    Plotar a distribuição desta população.

    Retirar amostras de tamanho n, calcular suas médias e plotar a distribuição amostral.



```python
from scipy.stats.distributions import randint
import random
```

```python
populacao = []

# seleciona 10000 números inteiros do intervalo [10, 100]
for i in range(10000):
    elemento = random.randint(10, 100)
    populacao.append(elemento)

# imprime
print(f"A população é: {populacao}")
```

```python
# Calcula a média da população
import statistics as stats
média_pop = stats.mean(populacao)
```

```python
# Calcula a variância das amostras
var_pop = stats.variance(populacao)
```

```python
print(f"A população tem média = {round(média_pop, 2)} e variância = {round(var_pop, 2)}")
```
